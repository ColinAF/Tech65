```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
url = 'https://github.com/mattharrison/datasets/raw/refs/heads/master/data/ames-housing-dataset.zip'
housing = pd.read_csv(url)
```

```python
housing.info()
```

```python
housing.dtypes # Display the data type in each column
```

```python
# display colunms with missing data in %
((housing
  .isna() # Not a number 
  .mean() # Take the mean
  * 100   # Convert to %
 )
  .pipe(lambda ser : ser[ser > 0]) # List all non-zero
  .sort_values(ascending=False) # Sort the values
)
```

```python
housing.Neighborhood.value_counts().plot.bar() # Houses per neighborhood
```
